```python
import pandas as pd
```

```python
data = pd.read_csv("/content/ML_Find_S_1.csv")
data
```

```python
def Find_S(d):
    columns = list(data.keys())
    col = columns[1:-1]
    h = ['O']*len(col)
    for i in range(len(d)):
        test = data[data['S.No']==i+1]
        if test['EnjoySport'][i]=='Yes':
            k=0
            for j in col:
                if h[k]!='O' and test[j][i]!=h[k]:
                    h[k]='?'
                elif h[k]=='O':
                    h[k]=test[j][i]
                k=k+1
    return h

h = Find_S(data)
print(h)
```

```python
def Test(t, h):
    h = [x.lower() for x in h]
    t = [x.lower() for x in t]
    chk=0
    for i in range(len(h)):
        if h[i]=='?':
            continue
        elif h[i]!=t[i]:
            chk=1
            break
    if chk==0:
        print("Given test case is in hypothesis")
    else:
        print("Given test case is not in hypothesis")


test = list(map(str, input("Enter test case : ").split()))

Test(test, h)
```
