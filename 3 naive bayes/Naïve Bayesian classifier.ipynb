```python
data = pd.read_csv("/content/dataset.csv")
data = pd.read_csv("dataset.csv")
data
```

```python
n = int(input("Train and test split : "))

train = data.iloc[:n]
test = data.iloc[n:]
```

```python
def Naive_Bayes(data):
    headings = list(data.keys())
    main_head = headings[1:-1]
    clas = list(data.iloc[:][headings[-1]])
    main_p = {x:clas.count(x)/len(data) for x in set(clas)}
    prob = {}
    for i in main_head:
        uniq = set(list(data[i]))
        prob[i] = {}
        for j in uniq:
            prob[i].update( {j: {x:0 for x in set(clas)}})

    for i in range(len(data)):
        for j in set(clas):
            if data.iloc[i][-1]==j:
                row = data.iloc[i]
                for k in main_head:
                    prob[k][row[k]][j]+=1/clas.count(j)
    return prob, main_p, list(set(clas))


train_result, main_p, clas = Naive_Bayes(train)
train_result
```

```python
def testing(data, main_p, clas, train_result):
    headings = list(data.keys())
    main_head = headings[1:-1]
    for i in range(len(data)):
        chances = {x:main_p[x] for x in clas}
        row = list(data.iloc[i])[1:-1]
        index = 0
        for j in row:
            for k in clas:
                chances[k]*=train_result[main_head[index]][j][k]
            index+=1
        print(chances)
        res = ""
        r = 0
        for a in chances.keys():
            if chances[a]>=r:
                res=a
                r=chances[a]
        # print("Therefore "+str(row)+" is "+res)
        print("Therefore "+data.iloc[i][0]+" is "+res)

testing(test, main_p, clas, train_result)
```
